```python
import pandas as pd
# business 데이터 불러오기
business_raw = pd.read_csv(r"C:\Users\seonu\Documents\ewha-marketing_research\datasets\yelp_dataset\yelp_business.csv")
```

```python
data_info = business_raw.info()
print(f"데이터 정보: {data_info}")
print("\n--- state 칼럼 ---")
unique_values_s= business_raw['state'].unique()
value_counts_s= business_raw['state'].value_counts()
print(f"\n고유값 목록: {unique_values_s}")
print("\n--- 빈도수 ---")
print(value_counts_s)
```

```python
# 카테고리 분해 후 전체 카테고리 목록
all_cats = business_raw['categories'].dropna().str.split(';').explode().str.strip()
cat_counts = all_cats.value_counts()
cat_unique= all_cats.unique()
print(f"전체 카테고리 수: {len(cat_counts)}개")
print("\n상위 50개 카테고리:")
print(cat_counts.head(50))
```

```python
# 고유 카테고리 전체 출력
for cat in cat_unique:
    print(cat)
    print(len(cat))
```

```python
# 숙박 관련 카테고리 검색
hotel_cats = cat_counts[cat_counts.index.str.contains('Hotel|Motel|Resort|Inn|Lodge|Hostel|Bed', case=False, na=False)]
print(hotel_cats)
```
