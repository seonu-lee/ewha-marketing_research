```python
from pandas import pandas as pd
# business 데이터 불러오기
business_raw = pd.read_csv(r"C:\Users\seonu\Documents\ewha-marketing_research\datasets\yelp_dataset\yelp_business.csv")

# 카테고리 분해 후 전체 카테고리 목록
all_cats = business_raw['categories'].dropna().str.split(';').explode().str.strip()
cat_counts = all_cats.value_counts()

print(f"전체 카테고리 수: {len(cat_counts)}개")
print("\n상위 50개 카테고리:")
print(cat_counts.head(50))
```

```python
# 숙박 관련 카테고리 검색
hotel_cats = cat_counts[cat_counts.index.str.contains('Hotel|Motel|Resort|Inn|Lodge|Hostel|Bed', case=False, na=False)]
print(hotel_cats)
```
